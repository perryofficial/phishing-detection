# **ML Model Development**

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
import sklearn.metrics as metrics
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, classification_report
import pickle
import gzip
```

```python
phishing_dataset = pd.read_csv("ML_Dataset/Dataset.csv")
phishing_dataset
# 0 = legitimate, 1 = phishing
```

```python
print(phishing_dataset['Type'].value_counts())
phishing_dataset['Type'].value_counts().plot(kind='pie', autopct='%1.0f%%',
                                               figsize=(6,6), fontsize=12)
```

```python
phishing_dataset.info()
```

```python
# Drawing correlation matrix
plt.figure(figsize=(50, 20))
sns.heatmap(phishing_dataset.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
labelled_bad = phishing_dataset[phishing_dataset['Type'] == 1]      # all phishing tuples
labelled_good = phishing_dataset[phishing_dataset['Type'] == 0]     # all legitimate tuples
# Random sampling
sampled_labelled_bad = labelled_bad.sample(n=55000, random_state=42)
sampled_labelled_good = labelled_good.sample(n=55000, random_state=42)
selected_data = pd.concat([sampled_labelled_bad, sampled_labelled_good])
selected_data = selected_data.sample(frac=1, random_state=42).reset_index(drop=True)
selected_data
```

```python
selected_data['Type'].value_counts().plot(kind='pie', autopct='%1.0f%%',
                                          figsize=(6,6), fontsize=12)
```

```python
X = selected_data.drop('Type', axis=1)
Y = selected_data['Type']
```

```python
clsf = ExtraTreesClassifier()
clsf.fit(X,Y)
print(clsf.feature_importances_)
```

```python
plt.figure(figsize = (12,8))
feat_importances = pd.Series(clsf.feature_importances_, index=X.columns)
feat_importances.nlargest(20).plot(kind='barh')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 42)
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto','sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
# Random search of parameters, using 5 fold cross validation, search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = RandomForestClassifier(), param_distributions = random_grid,
                               scoring='accuracy', n_iter = 10, cv = 5,
                               verbose=1, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
print("Model created successfully!")
```

```python
rf_random.best_params_
```

```python
prediction = rf_random.predict(X_test)
```

```python
labels = ["Not Phishing", "Phishing"]
cm = confusion_matrix(y_test, prediction)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot();
```

```python
accuracy = metrics.accuracy_score(y_test, prediction)
recall = metrics.recall_score(y_test, prediction)
precision = metrics.precision_score(y_test, prediction)
f1_score = metrics.f1_score(y_test, prediction)
```

```python
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1 Score:', f1_score)
print('MAE: ', metrics.mean_absolute_error(y_test,prediction))
print('MSE: ', metrics.mean_squared_error(y_test,prediction))
```

```python
rep1 = classification_report(y_test, prediction)
print(rep1)
```

```python
with gzip.open('phishing_model_rf.pkl.gz', 'wb') as file:
    pickle.dump(rf_random, file)
print(".pkl.gz file successfully created!")
```

```python
X4 = phishing_dataset.drop('Type', axis=1)
Y4 = phishing_dataset['Type']
```

```python
new_rf_forest2 = RandomForestClassifier(n_estimators=10)

# fit the model
new_rf_forest2.fit(X_train4,y_train4)
print("Model created successfully!")
```

```python
prediction_test4 = new_rf_forest2.predict(X_test4)
prediction_train4 = new_rf_forest2.predict(X_train4)
```

```python
cm4 = confusion_matrix(y_test4, prediction_test4)
disp4 = ConfusionMatrixDisplay(confusion_matrix=cm4, display_labels=["Not Phishing","Phishing"])
disp4.plot();
```

```python
accuracy_train2 = metrics.accuracy_score(y_train4, prediction_train4)
recall_train2 = metrics.recall_score(y_train4, prediction_train4)
precision_train2 = metrics.precision_score(y_train4, prediction_train4)
f1_score_train2 = metrics.f1_score(y_train4, prediction_train4)
accuracy_test2 = metrics.accuracy_score(y_test4, prediction_test4)
recall_test2 = metrics.recall_score(y_test4, prediction_test4)
precision_test2 = metrics.precision_score(y_test4, prediction_test4)
f1_score_test2 = metrics.f1_score(y_test4, prediction_test4)
```

```python
print('Training Accuracy:', accuracy_train2)
print('Testing Accuracy:', accuracy_test2)
print()
print('Training Precision:', precision_train2)
print('Testing Precision:', precision_test2)
print()
print('Training Recall:', recall_train2)
print('Testing Recall:', recall_test2)
print()
print('Training F1 Score:', f1_score_train2)
print('Testing F1 Score:', f1_score_test2)
```

```python
rep4 = classification_report(y_test4, prediction_test4)
print(rep4)
```

```python
training_accuracy = []
test_accuracy = []

depth = range(1,30)
for n in depth:
    forest_test = RandomForestClassifier(n_estimators=n)

    forest_test.fit(X_train4, y_train4)
    # record training set accuracy
    training_accuracy.append(forest_test.score(X_train4, y_train4))
    # record generalization accuracy
    test_accuracy.append(forest_test.score(X_test4, y_test4))

#plotting the training & testing accuracy for n_estimators from 1 to 30
plt.figure(figsize=None)
plt.plot(depth, training_accuracy, label="Training accuracy")
plt.plot(depth, test_accuracy, label="Test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_estimators")
plt.legend();
```
